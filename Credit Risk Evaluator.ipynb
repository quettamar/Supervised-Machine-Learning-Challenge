```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data
X_train = train_df.drop("loan_status", axis=1)
y_train = train_df['loan_status'].values
X_test = test_df.drop("loan_status", axis=1)
y_test = test_df['loan_status'].values
```

```python
X_train.info()
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X_dummies = pd.get_dummies(X_train)
print(X_dummies.columns)
X_dummies
```

```python
test_dummies = pd.get_dummies(X_test)
test_dummies
```

```python
y_test = pd.get_dummies(y_test)
y_train = pd.get_dummies(y_train)
```

```python
y_train = y_train.drop('low_risk', axis=1)
y_test = y_test.drop('low_risk', axis=1)
y_test
```

```python
checkpoint = X_dummies.copy()
```

```python
# add missing dummy variables to testing set
train_cols = list(X_dummies.columns)
test_cols = list(test_dummies.columns)
cols_not_in_test = {c:0 for c in train_cols if c not in test_cols}
test = test_dummies.assign(**cols_not_in_test)
test
```

```python
test = test.drop(['Unnamed: 0', 'index'], axis=1)
test_dummies = test_dummies.drop(['Unnamed: 0', 'index'], axis=1)
```

```python
X_dummies = X_dummies.drop(['Unnamed: 0', 'index'], axis=1)
```

```python

# I'm still trying to learn how the different models work but I  think a 
#random forest classifier would be a better fit for this data due to the 
# multiple features(?) that go into making the prediction. If it were a more
# simple dataset, a logistic regression would be better.
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(max_iter=750)
classifier
classifier.fit(X_dummies, y_train)
print(f"Training Data Score: {classifier.score(X_dummies, y_train)}")
print(f"Testing Data Score: {classifier.score(test, y_test)}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_dummies, y_train)
print(f'Training Score: {clf.score(X_dummies, y_train)}')
print(f'Testing Score: {clf.score(test, y_test)}')
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_dummies)
X_train_scaled = scaler.transform(X_dummies)
X_train_scaled
X_test_scaled = scaler.transform(test)
X_test_scaled
```

```python
#after scaling the data, I'm not sure which one will perform better because I'm still trying to figure out what scaling the data actually does. I'm thinking the logistic
#regression might do better this time around because all of the data will be equally important while the random forest classifies relies more on certain features being weighed
#more, making classifications easier.
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier = LogisticRegression(max_iter=500)
classifier
classifier.fit(X_train_scaled, y_train)
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

```python
# Which model performed better? How does that compare to your prediction? Write down your results and thoughts. 
# Once the data was scaled, the logistic regression actually performed a lot better than I thought it would. I was still under the impression that because there were 
# multiple features, the random forest classifier would perform better. I need to do more research into what scaling the data does.

#These last few cells I wasnt sure if we needed to do because the testing scores off the scaled data were already printed above but I thought it would be cool
#to see how the predictions would change.
```

```python
from sklearn.feature_selection import SelectFromModel
sel = SelectFromModel(clf)
sel.fit(X_train_scaled, y_train)
```

```python
X_selected_train = sel.transform(X_train_scaled)
X_selected_test = sel.transform(X_test_scaled)
```

```python
clf = LogisticRegression(max_iter=500).fit(X_selected_train, y_train)
print(f'Training Score: {clf.score(X_selected_train, y_train)}')
print(f'Testing Score: {clf.score(X_selected_test, y_test)}')
```

```python
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_selected_train, y_train)
print(f'Training Score: {clf.score(X_selected_train, y_train)}')
print(f'Testing Score: {clf.score(X_selected_test, y_test)}')
```
